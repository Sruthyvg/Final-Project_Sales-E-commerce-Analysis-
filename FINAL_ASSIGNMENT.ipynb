Why We Do This?

Derived columns help uncover deeper insights.

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

**# Convert categorical columns to category**

```python
cat_cols = ['Category','Sub-Category','Segment','Region','Ship Mode']
for col in cat_cols:
    df[col] = df[col].astype('category')
```

**# Profit Margin**

```python
df['Profit_Margin'] = df['Profit'] / df['Sales']
```

**# Year and Month from Order Date**

```python
df['Order_Year'] = df['Order Date'].dt.year
df['Order_Month'] = df['Order Date'].dt.month
df['Order_Month_Name'] = df['Order Date'].dt.strftime('%b')
```

```python
df[['Order Date','Order_Year','Order_Month','Order_Month_Name','Profit_Margin']].head()
```

**2.5 Remove Outliers**

Superstore has extreme values in Sales and Profit.

```python
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1

df = df[~((df['Sales'] < (Q1 - 1.5 * IQR)) | (df['Sales'] > (Q3 + 1.5 * IQR)))]
df.shape
```

**2.6 Filtering or Aggregating Data**


**2.6.1. Filter out negative profit rows**

```python
loss_orders = df[df['Profit'] < 0]
loss_orders.head()
```

**2.6.2 Filter high discount orders**

```python
high_discount = df[df['Discount'] > 0.5]
high_discount.head()
```

**2.6.3 Filter orders from a specific region**

```python
west_region = df[df['Region'] == 'West']
west_region.head()
```

**2.6.4 Total Sales by Category**

```python
df.groupby('Category', observed=True)['Sales'].sum()

```

**2.6.5 Total Profit by Region**

```python
df.groupby('Region', observed=True)['Profit'].sum()
```

**2.6.6 Average Discount by Sub‑Category**

```python
df.groupby('Sub-Category', observed=True)['Discount'].mean()
```

**2.6.7 Orders per Segment**

```python
df.groupby('Segment', observed=True)['Order ID'].count()
```

**3.1.1 Sales Distribution**

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], bins=40, kde=True)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```



**3.1.2 Profit Distribution**

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Profit'], bins=40, kde=True, color='orange')
plt.title('Distribution of Profit')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

**3.1.3 Category Counts**

```python
plt.figure(figsize=(7,4))
sns.countplot(x='Category', data=df)
plt.title('Count of Orders by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

**3.1.4 Region Counts**

```python
plt.figure(figsize=(7,4))
sns.countplot(x='Region', data=df)
plt.title('Count of Orders by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

**Univariate Insights**

1. Dominant Categories

The ***Office Supplies category*** has the highest number of orders, significantly more than the other categories. ***Furniture and Technology*** have much lower and almost similar order counts. This indicates that Office Supplies is a high‑volume category, while Furniture and Technology generate fewer but typically higher‑value orders.

2. Dominant Regions

The ***West region*** records the highest number of orders, followed by the East and Central regions. The ***South region*** has the lowest order count. This suggests that the West is the strongest performing market, while the South shows comparatively weaker demand.

3. Sales Distribution

The distribution of Sales is ***right‑skewed***, meaning most transactions involve low sales amounts, with a long tail of higher sales values. Even after removing outliers using the IQR method, the skewness remains. This indicates that large sales transactions are rare compared to smaller ones.

4. Profit Distribution

Profit shows a ***highly skewed distribution***, with a sharp peak around zero. Most transactions generate very small profit, and there are several negative profit values, indicating loss‑making orders. This suggests the presence of heavy discounting, returns, or high shipping costs affecting profitability.

**3.2.1 Sales vs Profit**

**3.2.2 Sales by Category**

```python
sales_by_cat = df.groupby('Category', observed=True)['Sales'].sum().reset_index()

plt.figure(figsize=(7,4))
sns.barplot(x='Category', y='Sales', data=sales_by_cat)
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()

```

**3.2.3 Profit by Region**

```python
profit_by_region = df.groupby('Region', observed=True)['Profit'].sum().reset_index()

plt.figure(figsize=(7,4))
sns.barplot(x='Region', y='Profit', data=profit_by_region)
plt.title('Total Profit by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.tight_layout()
plt.show()
```

**3.2.4 Discount vs Profit**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Discount', y='Profit', data=df, alpha=0.6, color='red')
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

**Bivariate Insights**

1. Categories Driving the Most Sales  

Based on the total sales analysis, ***Office Supplies*** contributes the highest overall sales among all categories. ***Furniture and Technology*** generate lower total sales in comparison. This indicates that Office Supplies is the primary revenue‑driving category, likely due to high order volume even if individual order values are smaller.

2. Profit Comparison Across Regions  

The ***West region*** shows the highest total profit, followed by the East region. The Central region generates moderate profit, while the ***South region ***performs the weakest in terms of profitability. This suggests that regional performance varies significantly, with the West being the most profitable market and the South potentially facing issues such as low demand, higher discounts, or unprofitable product mixes.

3. Relationship Between Discount and Profit  

The scatter plot of Discount vs Profit shows a ***clear negative relationship***. As discount levels increase, profit tends to decrease. Higher discounts often lead to reduced margins, and in many cases, high‑discount transactions result in negative profit. This indicates that aggressive discounting strategies may be harming overall profitability.

**3.3.1 Correlation Heatmap (Numerical Features)**

```python
plt.figure(figsize=(10,6))
corr = df[['Sales', 'Quantity', 'Discount', 'Profit', 'Profit_Margin']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

**3.3.2 Pivot Table: Sales by Region and Category**

```python
pivot_sales = pd.pivot_table(df, values='Sales',
                             index='Region',
                             columns='Category',
                             aggfunc='sum',
                             observed=False)


plt.figure(figsize=(8,5))
sns.heatmap(pivot_sales, annot=True, fmt='.0f', cmap='Blues')
plt.title('Sales by Region and Category')
plt.tight_layout()
plt.show()
```

**3.3.3 Segment‑wise Performance**

```python
segmentsegment_perf = df.groupby('Segment', observed=False)[['Sales', 'Profit']].sum().reset_index()


plt.figure(figsize=(7,4))
sns.barplot(x='Segment', y='Sales', data=segment_perf)
plt.title('Total Sales by Segment')
plt.xlabel('Segment')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()

plt.figure(figsize=(7,4))
sns.barplot(x='Segment', y='Profit', data=segment_perf)
plt.title('Total Profit by Segment')
plt.xlabel('Segment')
plt.ylabel('Total Profit')
plt.tight_layout()
plt.show()
```

**3.3.4 Monthly Sales Trend**

```python
monthly_sales = df.groupby(['Order_Year', 'Order_Month_Name'])['Sales'].sum().reset_index()

plt.figure(figsize=(10,5))
sns.lineplot(x='Order_Month_Name', y='Sales', hue='Order_Year', data=monthly_sales, marker='o')
plt.title('Monthly Sales Trend by Year')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

**Multivariate Insights**

1. Strong Region × Category Combinations

The heatmap of Sales by Region and Category shows clear performance
patterns across regions.

The ***West region*** performs strongly across all categories, especially in Office Supplies and Technology.

The East region also shows high sales in Office Supplies and Furniture.

The Central region has moderate sales across categories, with no standout category.

The South region consistently records the lowest sales in all categories.

Overall, the strongest combinations are West × Office Supplies, West × Technology, and East × Office Supplies.

2. Segment Comparison: Sales and Profit Together  

When comparing segments on both Sales and Profit:

The ***Consumer segment*** leads in both total sales and total profit, making it the most valuable customer group.

The Corporate segment also contributes significantly to profit, despite lower sales than Consumer.

The Home Office segment shows the lowest sales and profit, indicating limited revenue potential compared to the other segments.

This suggests that Consumer and Corporate segments should be prioritised for marketing and retention strategies.

3. Seasonal Trends and Year‑on‑Year Changes  

The Monthly Sales Trend by Year reveals clear seasonal patterns:

Sales consistently peak around ***August and November***, indicating strong mid‑year and holiday‑season demand.

Sales dip in February across all years, showing a recurring low‑demand period.

Year‑on‑year trends show steady growth, with later years (2016–2017) generally achieving higher monthly sales than earlier years.

These patterns highlight predictable seasonality and overall upward sales momentum over time.
